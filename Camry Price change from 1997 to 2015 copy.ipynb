```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_excel('Camry.xlsx')
print(df)
```

```python
x = list(df['Year'])
y = list(df['Price'])
annotations=list(df['Price'])
```

```python
plt.scatter(x,y,s=20,edgecolors="black")
plt.plot(x,y)
plt.xlabel("year") 
plt.ylabel("Price")
plt.rcParams["figure.figsize"] = (20, 10)
plt.title("Excel sheet to Scatter Plot")
for i, label in enumerate(annotations):
    plt.text(x[i], y[i],label,horizontalalignment='right')
plt.show()
```
